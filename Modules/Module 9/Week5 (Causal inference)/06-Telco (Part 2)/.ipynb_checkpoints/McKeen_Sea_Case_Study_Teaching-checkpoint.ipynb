# McKeen Sea Case Study

Let's start by loading the data.

```python
# This loads 'pandas' a library that can help you do all sorts of data manipulation. 
import pandas as pd

# This code builds a data table (dataframe) using a Google Drive url to a CSV file.
url='https://drive.google.com/file/d/1AVRC3HFX9SMI701S-ubEt5aKtoIWLg8D/view?usp=sharing'
df1 = pd.read_csv('https://drive.google.com/uc?id=' + url.split('/')[-2])
print(df1.columns.values)
```

Here's the performance of people that attended vs. people that did not.

```python
import numpy as np
import matplotlib.pyplot as plt

# Get confidence intervals
stds = df1.groupby("AttendedTraining").Performance.std()
counts = df1.groupby("AttendedTraining").Performance.count()
intervals = 1.96 * stds / np.sqrt(counts)

# Plot results
df1.groupby("AttendedTraining").Performance.mean().plot.bar(yerr=intervals, figsize=(8,7), capsize=5)
plt.ylabel("Avg. Performance")
plt.xticks(rotation=0)
plt.show()
```

What if we consider the past performance of people that attended the training?

```python
# Some auxiliar functions for plotting
def plot_avg_performance(data, groups=None):
    plot_auxiliar(data, groups, True)
    
def plot_number_of_employees(data, groups=None):
    plot_auxiliar(data, groups, False)

def plot_auxiliar(data, groups, mean=True):
    grouped = data.groupby(groups)
    rows = int(grouped.ngroups/2)
    fig, axs = plt.subplots(figsize=(16,7*rows), nrows=rows, ncols=2, gridspec_kw=dict(hspace=0.4))
    targets = zip(grouped.groups.keys(), axs.flatten())
    for i, (key, ax) in enumerate(targets):
        if mean:
            stds = grouped.get_group(key).groupby('AttendedTraining').Performance.std()
            counts = grouped.get_group(key).groupby('AttendedTraining').Performance.size()
            intervals = 1.96 * stds / np.sqrt(counts)
            performances = grouped.get_group(key).groupby('AttendedTraining').Performance.mean() 
            performances.plot(kind='bar', rot=0, ax=ax, yerr=intervals, capsize=5)
            ax.set_ylabel('Avg. performance')
            ax.set_ylim((0, 1))
        else:
            grouped.get_group(key).groupby('AttendedTraining').Performance.size().plot(kind='bar', rot=0, ax=ax)
            ax.set_ylabel('# of employees')
        key_s = []
        if type(key) is not tuple:
            key = (key,)
        for j, elem in enumerate(key):
            key_s.append(groups[j] + ' ' + elem )
        ax.set_title(' | '.join(key_s))

plot_avg_performance(df1, groups=['PastPerformance'])
```

Now it seems that the training hurt the performance for people with high AND low past performance! How is that possible? 

```python
plot_number_of_employees(df1, groups=['PastPerformance'])
```

Employees with high past performance were more likely to enroll in the training program, which would explain why people that enrolled in the program had a higher performance. However, once we account for past performance, we can see that people that took the training generally performed worse than people than didn't.

So, should we deploy the program? Do you feel more confident with your answer?

Before making a final decision, let's consider employee ternure as well.

```python
plot_avg_performance(df1, groups=['PastPerformance', 'RecentlyHired'])
```

The tides have turned again! People that took the training perform the same or better in all sub-categories. Why?

```python
plot_number_of_employees(df1, groups=['RecentlyHired'])
```

Confounding bias once more. Recently hired employees are more likely to attend the training, but these individuals also have a worse performance, explaining the spurious negative effect we found before. However, once we account for employee tenure in our analysis, the effect of the training program seems to be generally positive.

So... Do you think we can tell whether the training program was effective with these data?

Not without making additional assumptions! More specifically, we would have to assume that there is no other omitted variable influencing both performance and program attendance (such as past performance or employee tenure in our earlier examples). If we cannot make this "no-confounding" assumption, then we cannot be certain of whether the difference in performance is due to the program or some other ommited variable(s). Unfortunately, this is an assumption that we cannot test with data, which is the main reason why causal inference tends to be so challenging: it is hard to tell whether we are comparing apples with apples here.

For example, in this case, there is another variable excluded from our analysis: job satisfaction. By including this variable, we can see that the training program does not seem to have an effect on performance.

```python
import statsmodels.api as sm

features = ['AttendedTraining', 'PastPerformance', 'RecentlyHired', 'JobSatisfaction']
X = df1[features].copy()
for f in features:
  X[f] = (X[f] == "Yes") | (X[f] == "High")
y = df1['Performance']
X = sm.add_constant(X)
model1 = sm.OLS(y, X.astype(int))
model1 = model1.fit()
print(model1.summary())
```

This is an example of Simpson's paradox. Simpsons paradox refers to a phenomena whereby the association between a pair of variables (X, Y) reverses sign upon conditioning of a third variable, Z, regardless of the value taken by Z. If we partition the data into subpopulations, each representing a specific value of the third variable, the phenomena appears as a sign reversal between the associations measured in the disaggregated subpopulations relative to the aggregated data, which describes the population as a whole. So, what are the alternatives for McKeen Sea?
1. If the program is too expensive and risky, they could just decide not to launch it to other branches. After all, very often we need to make a decision with whatever data we have at hand (regardles off its limitations). And so, even though our assumptions might be violated, the observational data suggests that the program was not effective.

2. We could acquire additional data that does not suffer from confounding by running an experiment. Running such an experiment, however, could be costly and hard to design. For instance, how would you make sure that employees are randomly assigned to the training program? Would you force employees to attend the training program even when they do not want to? Similarly, would you stop employees from attending even if they want to?
